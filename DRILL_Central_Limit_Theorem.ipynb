```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pop1 = np.random.binomial(10, .2, 10000)
pop2 = np.random.binomial(10, .5, 10000)

print('Population one mean is {} '.format(pop1.mean()))
print('Population one standard deviation is {} '.format(pop1.std()))
print('Population two mean is {} '.format(pop2.mean()))
print('Population two standard deviation is {} '.format(pop2.std()))

plt.hist(pop1, alpha=.5, label='Population 1')
plt.hist(pop2, alpha=.5, label='Population 2')
plt.legend(loc='upper right')

plt.axvline(pop1.mean(), color='b',linestyle='solid', linewidth=2)
plt.axvline(pop1.mean() + pop1.std(), color = 'b', linestyle='dashed', linewidth=2)
plt.axvline(pop1.mean() - pop1.std(), color='b', linestyle='dashed', linewidth=2)

plt.axvline(pop2.mean(), color='g',linestyle='solid', linewidth=2)
plt.axvline(pop2.mean() + pop2.std(), color = 'g', linestyle='dashed', linewidth=2)
plt.axvline(pop2.mean() - pop2.std(), color='g', linestyle='dashed', linewidth=2)

plt.show()
```

```python
#Sample mean and standard deviation are relatively close to the population mean and std

sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

diff=sample2.mean() - sample1.mean()

print('Sample one mean is {} '.format(sample1.mean()))
print('Sample one standard deviation is {} '.format(sample1.std()))
print('Sample two mean is {} '.format(sample2.mean()))
print('Sample two standard deviation is {} '.format(sample2.std()))
print('The difference between the two samples is {}'.format(diff))

plt.hist(sample1, alpha=.5, label='sample 1')
plt.axvline(sample1.mean(), color='b',linestyle='solid', linewidth=2)
plt.axvline(sample1.mean() + sample1.std(), color = 'b', linestyle='dashed', linewidth=2)
plt.axvline(sample1.mean() - sample1.std(), color='b', linestyle='dashed', linewidth=2)

plt.hist(sample2, alpha=.5, label = 'sample 2')
plt.axvline(sample2.mean(), color='g', linestyle='solid', linewidth=2)
plt.axvline(sample2.mean() + sample2.std(), color='g', linestyle='dashed', linewidth=2)
plt.axvline(sample2.mean() - sample2.std(), color='g', linestyle='dashed',linewidth=2)
plt.legend(loc='upper right')

plt.show()
```

```python
# With sample size of 1000, the mean and the standard deviation are very close the population mean and std 

sample1_new = np.random.choice(pop1, 1000, replace=True)
sample2_new = np.random.choice(pop2, 1000, replace=True)

diff_new=sample2_new.mean() - sample1_new.mean()

print('Sample one mean is now {} '.format(sample1_new.mean()))
print('Sample one standard deviation is now {} '.format(sample1_new.std()))
print('Sample two mean is now {} '.format(sample2_new.mean()))
print('Sample two standard deviation is now {} '.format(sample2_new.std()))
print('The difference between the two samples is {}'.format(diff_new))

plt.hist(sample1_new, alpha=.5, label='sample 1_new')
plt.axvline(sample1_new.mean(), color='b',linestyle='solid', linewidth=2)
plt.axvline(sample1_new.mean() + sample1_new.std(), color = 'b', linestyle='dashed', linewidth=2)
plt.axvline(sample1_new.mean() - sample1_new.std(), color='b', linestyle='dashed', linewidth=2)

plt.hist(sample2_new, alpha=.5, label = 'sample 2_new')
plt.axvline(sample2_new.mean(), color='g', linestyle='solid', linewidth=2)
plt.axvline(sample2_new.mean() + sample2_new.std(), color='g', linestyle='dashed', linewidth=2)
plt.axvline(sample2_new.mean() - sample2_new.std(), color='g', linestyle='dashed',linewidth=2)
plt.legend(loc='upper right')

plt.show()
```

```python
#With the sample size of 20, the mean and the standard deviation are not as close to the mean and std of the population 
# as they were with larger sample size

sample1_small = np.random.choice(pop1, 20, replace=True)
sample2_small = np.random.choice(pop2, 20, replace=True)

diff_small=sample2_small.mean() - sample1_small.mean()


print('Sample one mean is now {} '.format(sample1_small.mean()))
print('Sample one standard deviation is now {} '.format(sample1_small.std()))
print('Sample two mean is now {} '.format(sample2_small.mean()))
print('Sample two standard deviation is now {} '.format(sample2_small.std()))
print('The difference between the two samples is {}'.format(diff_small))

plt.hist(sample1_small, alpha=.5, label='sample 1_small')
plt.axvline(sample1_small.mean(), color='b',linestyle='solid', linewidth=2)
plt.axvline(sample1_small.mean() + sample1_small.std(), color = 'b', linestyle='dashed', linewidth=2)
plt.axvline(sample1_small.mean() - sample1_small.std(), color='b', linestyle='dashed', linewidth=2)

plt.hist(sample2_small, alpha=.5, label = 'sample 2_small')
plt.axvline(sample2_small.mean(), color='g', linestyle='solid', linewidth=2)
plt.axvline(sample2_small.mean() + sample2_small.std(), color='g', linestyle='dashed', linewidth=2)
plt.axvline(sample2_small.mean() - sample2_small.std(), color='g', linestyle='dashed',linewidth=2)
plt.legend(loc='upper right')

plt.show()
```
