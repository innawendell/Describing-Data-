```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
%matplotlib inline
```

```python
df = pd.read_csv('C:\Inna\'s files\data\Sanctions\countries.csv')
x = df.loc[:, 'country_code'].value_counts()
#x.to_csv('C:\Inna\'s files\data\Sanctions\inna2.csv')
```

The data for this project was obtained from *OpenSantions* from the url: https://www.opensanctions.org/

This source is "a global database of persons and companies of political, criminal, or economic interest. It combines the most important sanctions lists, databases of politically exposed persons, and other public information into a single, easy-to-access dataset."

The following dataset shows top twenty **nationalities** (by country code) of the sanctioned persons, with Great Britain being the high outlier:

```python
df2 = pd.read_csv('C:\Inna\'s files\data\Sanctions\inna2.csv')
#print(df2)
new_df = df2.loc[lambda df: df2['frequency'] > 139, :]
print(new_df)

plt.scatter(x=new_df['country'], y = new_df['frequency'], color='green')
plt.xticks(rotation=45)
plt.title('Frequency by Nationality')
```

The next dataset shows top twenty **birthplaces** of the sanctioned persons. 

The most interesting finding from this dataset is that Great Britain, which is by far the most common nationality of the santioned persons, does not make it top twenty birthplaces. The high outlier and the most frequent birthplace is Afghanistan. Great Britain appears to be the top destination for persons who end up on the sanctions lists.

```python
df3 = pd.read_csv('C:\Inna\'s files\data\Sanctions\inna3.csv')
new_df1 = df3.loc[lambda df: df2['frequency'] > 140, :]
print(new_df1)

plt.scatter(x=new_df1['country'], y=new_df1['frequency'], color='purple')
plt.xticks(rotation=45)
plt.title('Frequency by Birthplace')
```

The following visualizations compare side by side the nationalities and birthplaces of the sanctioned persons. While the overlap is clear, the high outliers vary significantly. Additionally, Russia makes it top 20 **birthplaces** and only top 10 **nationalities** of the sanctioned persons.

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.scatter(x=new_df['country'], y = new_df['frequency'], color='green')
plt.xticks(rotation=45)
plt.title('Frequency by Nationality')

plt.subplot(1, 2, 2)
plt.scatter(x=new_df1['country'], y=new_df1['frequency'], color='purple')
plt.xticks(rotation=45)
plt.title('Frequency by Birthplace')     

plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.boxplot(new_df['frequency'])
plt.title('Nationalities')


plt.subplot(1, 2, 2)
plt.boxplot(new_df1['frequency'])
plt.title('Birthplaces')

plt.tight_layout()
plt.show()
```
